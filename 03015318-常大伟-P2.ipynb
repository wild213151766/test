## （一）问题描述

###       装有两个给水加热器的再热回热循环

**例 8.6 : ** 在使用两台给水加热器——一台闭式给水加热器和一台开式给水加热器的再热回热循环中，

   * 8.0 MPa, 480°C的水蒸气进入低压汽轮机，等熵膨胀至0.7MPa。 
   

   * 水蒸气被加热到 440°C，之后进入下一级汽轮机，并等熵膨胀至0.008MPa。
   

   * 从低压汽轮机中抽出 y' Kg 压力为 2 MPa的水蒸气，输送到闭式给水加热器。
   

   * 从闭式给水加热器离开的给水为 205°C、8.0 MPa，y1' Kg 压力为 2 MPa的水蒸气冷凝为压力为 2 MPa的饱和水。
   

   * y' Kg 压力为 2 MPa冷凝饱和水通过阀门等焓进入开式给水加热器。 
   

   * 从高压汽轮机中抽出y''Kg压力为0.3 MPa的水蒸气，输送到工作压力为0.3 MPa的开式给水加热器。在开式给水加热器中的水蒸气冷凝为0.3 MPa的饱和水。 
   
 
   * 该循环的净功率输出为 100 MW。
   

   * 完全封闭，没有任何热量到周围环境。
   

   * 当工质经过汽轮机、水泵、锅炉、再热器和冷凝器时可逆。
   


**已知:** 再热回热循环以蒸汽为工质， 部分状态点的压力和温度已给出，并已知净功率输出为 100 MW。
   

**求解:** 通过已知状态点的压力等信息，计算出13个状态点的物理性质，进而计算循环的热效率和进入低压汽轮机的质量流量，单位为Kg/h。

## （二）设计思路与模块说明


### 1.整体设计思路

  
   * **(a) 状态点计算（函数CalState())**：
   
   
       通过各个状态点之间的关系，从已知状态点的相关数据计算出该循环中13个状态点的数据，包括压力、温度、焓值、熵值、比容和干度等。
   
   
   * **(b) 循环过程中的计算(函数CalCycle(h,Wcycledot))**：

       算出工质在整个循环中的净做功量和净吸热量，从而得出循环的热效率和总质量流量。
   
   
   * **(c) 各个点质量流量的计算（函数FMDOT(results))**：
   
   
       通过(b)中计算所得的y'、y''值和总质量流量，计算得出各个状态点的质量流量。
   
   
   * **(d) 计算结果输出（函数Print (p,t,h,s,v,x,cycleResults,Wcycledot,cycleFDOT,cycleMDOT))**：
       
       
       将上述计算结果清晰地输出。
   
   
   * **(e) 图像表示（函数Plt(t,s))**：
       
       
       将循环过程通过图像表示出来。
       
       
   以上各部分的计算通过函数封装，在主体部分调用函数，实现功能。
   
       
       

###  2.各模块设计思路和说明

**(a) 状态点计算（函数CalState())**：


*  函数CalState通过各个状态点之间的关系，从已知状态点的相关数据计算出该循环中13个状态点的数据。

  
*  函数CalState结合（一）问题描述中的分析并调用seuif97中的相关函数，已知压力、温度、焓值、熵值、比容和干度中的两个值，求出其他四个值。

  
*  每一个状态点的压力（p/MPa)、温度（t/℃）、焓值（h/KJ/㎏）、熵值（s/KJ/㎏）、比容（v/m³/㎏）和干度（x）六个数值用列表（List）储存，通过序列


    号访问数据。


代码如下：
  

```python
from seuif97 import *


# 计算状态点 1~13 的压力（p/MPa)、温度（t/℃）、焓值（h/KJ/㎏）、熵值（s/KJ/㎏）、比容（v/m³/㎏）和干度（x）六个值

def CalState():
    
    p=[]                      # 压力（p/MPa)
    t=[]                      #温度（t/℃）
    h=[]                      #焓值（h/KJ/㎏）
    s=[]                      #熵值（s/KJ/㎏）
    v=[]                      #比容（v/m³/㎏）
    x=[]                      #干度（x）


    numNodes = 13             #预留空间
    i = 0
    while (i<numNodes): 
        p.append(None)
        t.append(None)
        h.append(None)
        s.append(None)
        v.append(None)
        x.append(None)
        i=i+1


    # State  1
    p[0] = 8.0                   #  in MPa
    t[0] = 480                   #  in ℃
    h[0] = pt2h(p[0],t[0])        
    s[0] = pt2s(p[0],t[0])
    v[0] = pt2v(p[0],t[0])
    x[0] = pt2x(p[0],t[0])         

    # State  2
    p[1] = 2.0                   #  in MPa
    s[1] = s[0]                  # state 1 to 2 为等熵过程
    t[1] = ps2t(p[1],s[1])
    h[1] = ps2h(p[1],s[1])
    x[1] = ps2x(p[1],s[1])
    v[1] = ps2v(p[1],s[1])

    # State  3
    p[2] = 0.7                   #  in MPa
    s[2] = s[0]                  # state 2 to 3 为等熵过程
    t[2] = ps2t(p[2],s[2])
    h[2] = ps2h(p[2],s[2])
    x[2] = ps2x(p[2],s[2])
    v[2] = ps2v(p[2],s[2])

    # State  4
    t[3] = 440                   #  in ℃
    p[3] = p[2]                  # state 3 to 4 为等压再热过程
    h[3] = pt2h(p[3],t[3])
    s[3] = pt2s(p[3],t[3])
    v[3] = pt2v(p[3],t[3])
    x[3] = pt2x(p[3],t[3])

    # State  5
    p[4] = 0.3                   #  in MPa
    s[4] = s[3]                  # state 4 to 5 为等熵过程
    t[4] = ps2t(p[4],s[4])
    h[4] = ps2h(p[4],s[4])
    v[4] = ps2v(p[4],s[4])
    x[4] = ps2x(p[4],s[4])

    # State  6
    p[5] = 0.008                 #  in MPa
    s[5] = s[3]                  # state 5 to 6 为等熵过程
    t[5] = ps2t(p[5],s[5])
    h[5] = ps2h(p[5],s[5])
    v[5] = ps2v(p[5],s[5])
    x[5] = ps2x(p[5],s[5])

    # State  7
    p[6] = p[5]                 # in MPa，state 6 to 7 为等压冷凝过程
    x[6] = 0                    # 冷凝饱和水，干度为0
    t[6] = px2t(p[6],x[6])
    h[6] = px2h(p[6],x[6])
    v[6] = px2v(p[6],x[6])
    s[6] = px2s(p[6],x[6])

    # State  8
    s[7] = s[6]                 # state 7 to 8 为等熵过程
    p[7] = p[8] =0.3            # in MPa，state 8 to 9为等压过程
    x[7] = ps2x(p[7],s[7])
    t[7] = ps2t(p[7],s[7])
    v[7] = ps2v(p[7],s[7])
    h[7] = ps2h(p[7],s[7])

    # State  9
    p[8] = 0.3                  # in MPa
    x[8] = 0                    # 饱和水，干度为0
    t[8] = px2t(p[8],x[8])
    h[8] = px2h(p[8],x[8])
    v[8] = px2v(p[8],x[8])
    s[8] = px2s(p[8],x[8])

    # State  10
    s[9] = s[8]                 # state 9 to 10 为等熵过程
    p[9] = p[10] =8.0           # in MPa，state 10 to 11为等压过程
    x[9] = ps2x(p[9],s[9])
    t[9] = ps2t(p[9],s[9])
    v[9] = ps2v(p[9],s[9])
    h[9] = ps2h(p[9],s[9])

    # State  11
    p[10] = 8.0                 # in MPa
    t[10] = 205                 #  in ℃
    h[10] = pt2h(p[10],t[10])        
    s[10] = pt2s(p[10],t[10])
    v[10] = pt2v(p[10],t[10])
    x[10] = pt2x(p[10],t[10])

    # State  12
    p[11] = 2.0                 # in MPa
    x[11] = 0                   # 饱和水，干度为0
    t[11] = px2t(p[11],x[11])
    h[11] = px2h(p[11],x[11])
    v[11] = px2v(p[11],x[11])
    s[11] = px2s(p[11],x[11])

    # State  13
    p[12] = 0.3                 # in MPa
    h[12] = h[11]               # state 12 to 13为等焓过程
    x[12] = ph2x(p[12],h[12])
    t[12] = ph2t(p[12],h[12])
    v[12] = ph2v(p[12],h[12])
    s[12] = ph2s(p[12],h[12])

    return (p,t,h,s,v,x)
```

**(b) 循环热效率计算（函数CalCycle(h,Wcycledot)）**：

* 循环的净功量计算公式为

    $\dot{W}_{net}=\dot{W}_t-\dot{W}_{p1}-\dot{W}_{p2}$
    

* 汽轮机做功与质量之比的计算公式为

    $\frac{\dot{W}_{t1}}{\dot{m}}=h_1-h_2$ 
    
    $\frac{\dot{W}_{t2}}{\dot{m}}=(1-y')\times(h_2-h_3)$ 
    
    $\frac{\dot{W}_{t3}}{\dot{m}}=(1-y')\times(h_4-h_5)$ 
    
    $\frac{\dot{W}_{t4}}{\dot{m}}=(1-y'-y'')\times(h_5-h_6)$ 
    
    $\frac{\dot{W}_t}{\dot{m}}=\dot{W}_{t1}+\dot{W}_{t2}+\dot{W}_{t3}+\dot{W}_{t4}$ 
    
    
* 水泵1、 2做功与质量之比的计算公式为

    $\frac{\dot{W}_{p1}}{\dot{m}}=(1-y'-y'')\times(h_8-h_7)$
    
    $\frac{\dot{W}_{p2}}{\dot{m}}=h_{10}-h_9$ 
    

    公式中的 m 为质量流量。 
  
  
* 该循环的吸热量为锅炉和再热器两部分

    锅炉：$\frac{\dot{Q}_1}{\dot{m}}=h_1-h_11$
    
    再热器：$\frac{\dot{Q}_2}{\dot{m}}=(1-y')\times(h_4-h_3)$
    
    总吸热量：$\frac{\dot{Q}_{in}}{\dot{m}}=\frac{\dot{Q}_1}{\dot{m}}+\frac{\dot{Q}_2}{\dot{m}}$
    

* 循环的热效率计算公式为

    $\eta=\frac{\dot{W}_{net}}{\dot{Q}_{in}}=\frac{\dot{W}_t-\dot{W}_{p1}-\dot{W}_{p2}}{\dot{Q}_{in}}$
    

* 进入汽轮机的质量流量（Kg/h）计算公式为

    $\dot{m}=\frac{\dot{Wcycledot}}{\dot{W}_{net}}\times3600000$
    
**该函数通过以上公式计算出结果，并储存在字典（Dictionary）results中，代码如下：**

```python
#计算循环各个部分做功和吸放热

def CalCycle(h,Wcycledot):

    results = {}
    #第一次从汽轮机抽出的蒸汽量
    y01 = results['y1'] = (h[10] - h[9])/(h[1]-h[11])
    
    #第二次从汽轮机抽出的蒸汽量
    y02 = results['y2'] = (h[8] - (1-y01)*h[7] - y01*h[12]) / (h[4]-h[7])

    #  汽轮机，Turbine
    #state 1 to 2
    results['Wt1'] = h[0] - h[1]
    #state 2 to 3
    results['Wt2'] = (1-y01 )*(h[1] - h[2])
    #state 3 to 4
    results['Qtadded'] = (1-y01 )*(h[3] - h[2])
    #state 4 to 5
    results['Wt3'] = (1-y01 )*(h[3] - h[4])
    #state 5 to 6
    results['Wt4'] = (1-y01-y02 )*(h[4] - h[5])
    #汽轮机总做功量
    results['Wt'] = results['Wt1']+results['Wt2']+results['Wt3']+results['Wt4']

    
    #  冷凝器，Condenser
    #state 6 to 7，发热量
    results['Qout'] = (1-y01-y02 )*(h[5] - h[6])

    
    #  水泵1，Pump1
    #state 7 to 8，水泵 1 的耗功量
    results['Wp1'] = (1-y01-y02 )*(h[7] - h[6])

    
    #  开式给水加热器，Open heater
    #state 8 to 9

    
    #  水泵2，Pump2
    #state 9 to 10，水泵 2 的耗功量
    results['Wp2'] = h[9] - h[8]

    
    #  闭式给水加热器，Closed heater
    #state 10 to 11

    
    #  阀门，Trap
    #state 12 to 13，等焓过程

    
    #  锅炉，Steam generator
    #state 11 to 1，锅炉吸热量
    results['Qin1'] = h[0] - h[10]

    
    #净功量
    results['Wnet'] = results['Wt']-results['Wp1']-results['Wp2']
    
    #总吸热量
    results['Qin']=results['Qin1']+results['Qtadded']
    
    #热效率
    results['Efficiency'] = results['Wnet'] / results['Qin']
    
    #质量流量
    results['Mass Flow'] = 3600*Wcycledot/results['Wnet']*10**3
    
    return results
```

**(c) 进入低压汽轮机时的质量流量计算（函数FMDOT(results)）**：

* 通过问题描述中的分析，可以得知各个状态点的流量在总质量流量中所占的百分数FDOT，通过以下公式计算求得该状态点的质量流量MDOT，并储存在列表中
  
  ${MDOT}={FDOT}\times{m}$
  
  
 代码如下：

```python
#计算各点的质量流量

def FMDOT(results):
    #FDOT and MDOT
    FDOT=[]
    MDOT=[]
    numNodes=13
    i=0

    while (i<numNodes): 
        FDOT.append(None)
        MDOT.append(None)
        i=i+1

    FDOT[0]=1
    MDOT[0]=FDOT[0]*results['Mass Flow']
    
    FDOT[1]=results['y1']
    MDOT[1]=FDOT[1]*results['Mass Flow']
    
    FDOT[2]=1-results['y1']
    MDOT[2]=FDOT[2]*results['Mass Flow']
    
    FDOT[3]=1-results['y1']
    MDOT[3]=FDOT[3]*results['Mass Flow']
    
    FDOT[4]=results['y2'] 
    MDOT[4]=FDOT[4]*results['Mass Flow']
    
    FDOT[5]=FDOT[6]=FDOT[7]=1-results['y1']-results['y2'] 
    MDOT[5]=MDOT[6]=MDOT[7]=FDOT[5]*results['Mass Flow']
    
    FDOT[8]=FDOT[9]=FDOT[10]=1
    MDOT[8]=MDOT[9]=MDOT[10]=FDOT[8]*results['Mass Flow']
    
    FDOT[11]=FDOT[12]=results['y1']
    MDOT[11]=MDOT[12]=FDOT[11]*results['Mass Flow']

    return (FDOT,MDOT)
```

**(d) 计算结果输出（函数Print (p,t,h,s,v,x,cycleResults,Wcycledot,cycleFDOT,cycleMDOT)）**：

* 输出计算结果，通过格式化，使数据输出清晰明了。

代码如下：

```python
#输出计算结果

def Print (p,t,h,s,v,x,cycleResults,Wcycledot,cycleFDOT,cycleMDOT):

    print ('{:>5}'.format('state'),'{:>10}'.format('P(MPa)'),'{:>10}'.format('T(°C)'),'{:>10}'.format('H(kJ/kg)'),
           '{:>10}'.format('S(kJ/kg.K)'),'{:>10}'.format('V(m³/kg)'),'{:>7}'.format('X'),'{:>10}'.format('FDOT'),
           '{:>18}'.format('MDOT(kg/h)'))
    i=0
    Num=13
    while (i<Num):
        print ('%4d'%(i+1),'%*.*f'%(10,3,p[i]),'%*.*f'%(10,3,t[i]),'%*.*f'%(10,3,h[i]),'%*.*f'%(10,3,s[i]),
               '%*.*f'%(10,3,v[i]),'%*.*f'%(10,3,x[i]),'%*.*f'%(10,4,cycleFDOT[i]),'%*.*f'%(15,3,cycleMDOT[i]))
        i=i+1

    print("\n  1.The net power output of the cycle: ", Wcycledot, "MW")
    print("\n  2.Efficiency: ", '%.3f' % (cycleResults['Efficiency']* 100), "%")
    print("\n  3.The mass flow rate: ",  '%.3f' % cycleResults['Mass Flow'], "kg/h")
```

**(e) 图像表示（函数Plt(t,s)）**：

* 将回热再热循环系统通过图像更清晰直观地描绘出来。

代码如下：

```python
#画图

from seuif97 import *
import matplotlib.pyplot as plt
import numpy as np


def Plt(t,s):
    npt = np.linspace(10,647.096-273.15,200)            # 温度范围
    svap = [s for s in [tx2s(t, 1) for t in npt]]
    sliq = [s for s in [tx2s(t, 0) for t in npt]]
    plt.plot(svap, npt, '#8B0000')
    plt.plot(sliq, npt, '#800080')
    t1=px2t(8.0,0.0)
    s1=px2s(8.0,0.0)
    t2=px2t(8.0,1.0)
    s2=px2s(8.0,1.0)

    # Rankine T-S

    tt=t[0:11]
    tt.append(t1)
    tt.append(t2)
    tt.append(t[0])
    ss=s[0:11]
    ss.append(s1)
    ss.append(s2)
    ss.append(s[0])

    plt.plot(ss, tt, 'g-')

    plt.xlabel('Entropy (kJ/(kg K)')
    plt.ylabel('Temperature (°C)')
    plt.grid()
    plt.show()
```

**（f）主函数：**

* Rankine循环的主函数，通过调用以上五个函数，完成状态点的计算、循环的计算，并把结果清楚明了地输出。

代码及结果如下：

```python
#朗肯循环

Wcycledot = 100.00     # in MW，循环的净功率

p,t,h,s,v,x = CalState()  #计算各状态点的值

cycleResults = CalCycle(h,Wcycledot)  #计算循环的热效率和总质量流量

cycleFDOT,cycleMDOT = FMDOT(cycleResults)  #计算各状态点的质量流量

Print (p,t,h,s,v,x,cycleResults,Wcycledot,cycleFDOT,cycleMDOT)  #输出结果

Plt (t,s)  #画图表示
```

**1.** 整个程序通过五个函数，包括**CalState**() 、 **CalCycle**(h,Wcycledot) 、 **FMDOT**(results) 、 **Print** (p,t,h,s,v,x,cycleResults,Wcycledot,cycleFDOT,cycleMDOT) 、 **Plt**(t,s)，来实现不同功能，最后在主函数**Rankine**中调用以上函数，实现回热再热循环的计算。

**2.** 用列表（List）储存状态点的压力、温度、焓值和干度等数值，并通过序列号访问数据；通过字典（Dictionary）储存计算循环热效率和质量流量过程中
的计算结果，数据结构字典包括键（Key）和值（Value），访问数据时可以通过Key来区分不同的结果，而不是像列表只能通过序列号来区分，代码更具可读性。

**3.** 计算状态点时，每一个状态点的六个数值的计算都需要写出相应代码，状态点较多时比较繁琐，阅读step3、4中示例代码后发现，使用类会更加方便。

**4.** 相比于C++，Python语言对数据类型的严苛程度更低，定义一个变量时不用指明其数据类型，而是由赋值给变量的对象来决定，更具灵活性。

**5.** 这次的练习先使用Visual Studio Code编写，后使用Notebook编辑。使用Visual Studio Code在运行和调试程序时更慢一些，使用Notebook相对快一些。
使Visual Studio Code时，不同的函数封装在不同的Python文件中，在主函数中调用需要导入；使用Notebook时，不同的函数在不同Cell中，但最后主函数调用时不需要导入，相当于整个代码在同一个Python文件中，这是Visual Studio Code和Notebook的些许不同。
